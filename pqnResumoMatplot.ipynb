```python
import numpy as np
```

```python
x=np.arange(0, 10, 0.1)
fig, ax = plt.subplots(ncols=2, nrows=2, figsize=(15,10))
fig.suptitle('Gr√°fico em conjunto', fontsize=30)

ax[0, 0].plot(np.sin(x))
ax[0, 0].set_title('Tabela 1')

ax[0, 1].plot(np.cos(x))
ax[0, 1].set_title('Tabela 2')

ax[1, 0].plot(np.tan(x))
ax[1, 0].set_title('Tabela 3')

ax[1, 1].plot(x)
ax[1, 1].set_title('Tabela 4')
```
