```python
chart = ctc.Bar('Médias Notas MS', width='500px', height='400px')#essa primeira linha, usa uma variável para receber o tipo e as dimnensoes, parecido com "plt.figure"
chart.set_options(
    labels=['C.Natureza', 'C.Humanas', 'Linguagens', 'Matemática'],
    x_label='Provas',
    y_label='Média por provas',
    colors=['#FFF1C9', '#F7B7A3', '#EA5F89', '#9B3192']
)
chart.add_series('Média', list(dados_ms[['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC',
       'NU_NOTA_MT']].mean()))
chart.render_notebook()
```

```python
radar_chart = ctc.Radar('Médias Notas MS', width='500px', height='400px')
radar_chart.set_options(
    labels=['C.Natureza', 'C.Humanas', 'Linguagens', 'Matemática'],
    is_show_legend=True, 
    legend_pos='upRight'
)
radar_chart.add_series('Média Estado', list(dados_ms[['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC',
       'NU_NOTA_MT']].mean()))
#para colocar um segundo radar, basta inserir um segundo "radar_chart.add_series"
radar_chart.render_notebook()
```
