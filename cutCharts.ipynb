```python
chart = ctc.Bar('M√©dias Notas MS', width='500px', height='400px')#essa primeira linha, usa uma vari√°vel para receber o tipo e as dimnensoes, parecido com "plt.figure"
chart.set_options(
    labels=['C.Natureza', 'C.Humanas', 'Linguagens', 'Matem√°tica'],
    x_label='Provas',
    y_label='M√©dia por provas',
    colors=['#FFF1C9', '#F7B7A3', '#EA5F89', '#9B3192']
)
chart.add_series('M√©dia', list(dados_ms[['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC',
       'NU_NOTA_MT']].mean()))
chart.render_notebook()
```

```python
#nao renderizou o gr√°fico
```

```python
radar_chart = ctc.Radar('M√©dias Notas MS', width='500px', height='400px')
radar_chart.set_options(
    labels=['C.Natureza', 'C.Humanas', 'Linguagens', 'Matem√°tica'],
    is_show_legend=True, 
    legend_pos='upRight'
)
radar_chart.add_series('M√©dia Estado', list(dados_ms[['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC',
       'NU_NOTA_MT']].mean()))
#para colocar um segundo radar, basta inserir um segundo "radar_chart.add_series"
radar_chart.render_notebook()
```
