```python
#apply, aplicar uma função a uma coluna
```

```python
def checar_dif_acao(diff):
  if diff > 0:
    return 'up'
  elif diff < 0:
    return 'down'
  else:
    return 'equal'
```

```python
acoes['label diff apply'] = acoes['diff'].apply(checar_dif_acao)
```
