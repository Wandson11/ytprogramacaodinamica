```python
import pandas as pd
```

```python
microdados = pd.read_csv('/content/drive/MyDrive/MICRODADOS_ENEM_2018.csv', encoding='latin-1', sep=';', nrows=1000000)
```

```python
microdados.head()
```

```python
microdados.columns.values
```

```python
#APURAR: QUAL É A MÉDIA EM CADA DISCIPLINA DO ENEM POR MUNICÍPIO?
```

```python
microdados_colunas_selecionadas = microdados.filter(items=['CO_MUNICIPIO_ESC', 'NO_MUNICIPIO_ESC', 'CO_UF_ESC', 'SG_UF_ESC', 'NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC',
       'NU_NOTA_MT', 'NU_NOTA_REDACAO'])
```

```python
microdados_colunas_selecionadas
```

```python
media_por_cidade = microdados_colunas_selecionadas.groupby(['CO_MUNICIPIO_ESC', 'NO_MUNICIPIO_ESC', 'CO_UF_ESC', 'SG_UF_ESC'], as_index=False).mean()
```

```python
media_por_cidade
```

```python
#os parametros dentro do groupby sao usados como um indice, assim, deixou de ser como uma coluna. para forçar a voltar a ser coluna
#o comando ou expressao é a: as_index=False
```

```python
#salvando o arquivo.
media_por_cidade.to_csv('mediacidadetocsv.csv')
```
