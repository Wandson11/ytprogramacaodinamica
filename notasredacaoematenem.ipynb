```python
novosdados['SG_UF_ESC'] = microdados.SG_UF_ESC
```

```python
#o procedimento acima, apenas acrescentou uma nova coluna ao df mais recente sem precisar de refazer todo o procedimento
```

```python
novosdados.head()
```

```python
#deu certo, agora foi feito um procedimento similar aos anteriores referente ao groupby e filter
novosdados.filter(items=['NU_NOTA_REDACAO', 'Q002 Convertido'])\
.where(novosdados.SG_UF_ESC == 'RJ')\
.groupby('Q002 Convertido')\
.mean().sort_values('NU_NOTA_REDACAO', ascending=False)
#AQUI foi usado o where para encontrar somente notas relacionadas ao estado do rj
```

```python
#resolvi fazero procedimento acima com ms
novosdados.filter(items=['NU_NOTA_REDACAO', 'Q002 Convertido'])\
.where(novosdados.SG_UF_ESC == 'MS')\
.groupby('Q002 Convertido')\
.mean().sort_values('NU_NOTA_REDACAO', ascending=False)
```

```python
#com sp
novosdados.filter(items=['NU_NOTA_REDACAO', 'Q002 Convertido'])\
.where(novosdados.SG_UF_ESC == 'SP')\
.groupby('Q002 Convertido')\
.mean().sort_values('NU_NOTA_REDACAO', ascending=False)
```

```python
# para nao precisar fazer um para cada estado, foi apresentado groupby com 2 termos.
novosdados.filter(items=['SG_UF_ESC', 'NU_NOTA_REDACAO', 'Q002 Convertido']).groupby(['SG_UF_ESC', 'Q002 Convertido'])\
.mean()
```

```python
novosdados.filter(items=['SG_UF_ESC', 'Q002 Convertido', 'NU_NOTA_REDACAO']).groupby(['Q002 Convertido', 'SG_UF_ESC'])\
.mean().plot()
```

```python
fig, ax = plt.subplots(figsize=(15,6))
plt.suptitle('Nota Redacao x Escolaridade x Estado')

novosdados.filter(items=['SG_UF_ESC', 'Q002', 'NU_NOTA_REDACAO'])\
.where(novosdados.Q002 != 'H')\
.groupby(['Q002', 'SG_UF_ESC'])\
.mean()\
.sort_values('SG_UF_ESC', ascending=False)\
.unstack().plot(ax= ax)
```
